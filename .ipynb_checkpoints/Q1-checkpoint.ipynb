# Design the model

## Dataset Splitting

```python
df.drop(columns=['Stamps','Time_order'], inplace=True)
```

```python
X = df.drop(columns="Fault")
y = df.Fault

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## Training

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
from jcopml.tuning import random_search_params as rsp
```

```python
preprocessor = ColumnTransformer([
    ('numeric', num_pipe(), ["CP1T1", "CP1T2", "CP2T1", "CP2T2", "CP1V1", "CP1V2", "CP2V1", "CP2V2"])
])

pipeline = Pipeline([
    ('prep', preprocessor),
    ('algo', RandomForestClassifier(n_jobs=-1, random_state=42))
])

model = RandomizedSearchCV(pipeline, rsp.rf_params, cv=4, n_jobs=-1, verbose=1)
model.fit(X_train, y_train)

print(model.best_params_)
print(model.score(X_train, y_train), model.best_score_, model.score(X_test, y_test))
```

## Feature Importance

```python
df_imp = mean_score_decrease(X_train, y_train, model, plot=True)
```

# Bonus (predicting program)

```python
pump_faulty_predictor = load_model("model/pump_faulty.pkl")
```

### Assuming that the data input is in json format
### Each timestamp send the data to machine to predict fault

```python
def predict_pump_faulty(sensor_data):
    X_pred = pd.DataFrame([sensor_data], columns=['CP1T1', 'CP1T2', 'CP2T1', 'CP2T2', 'CP1V1', 'CP1V2', 'CP2V1', 'CP2V2'])
    faulty = pump_faulty_predictor.predict(X_pred)[0]
    print("PUMP FAULT!") if faulty else print("Pump is fine")
    return faulty
```
